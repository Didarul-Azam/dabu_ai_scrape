```python
from yt_downloader import download_song
import asyncio
import os
```

```python
list_of_songs = [('Tunnel of Love','Dire Strait'),('Lonesome Town','Riky Nelson'),('As You Fade Away','Neffex'),('Runaway','Aurora'),('Closer','Chainsmoke'),('11 am','Incubus'),('High Hopes','Pink Floyd'),('Unforgiven 2','Metallica'),('Sultan of Swing','Dire Straits'),('Perfect','Ed Sheeran'),('Skin','Rag_n Bone Man')]
len(list_of_songs)
```

## Download Each song once in parallel
### with some deliberate misspell in artist name for Ricky Nelson and The chainsmokers, and Rag'N'Bone Man 

```python
tasks = []
for song_title,artist in list_of_songs:
    task = download_song(song_title,artist,f'{song_title}')
    tasks.append(task)
await asyncio.gather(*tasks)    
```

## Downloading one song 50 times in parallel

```python
os.makedirs('Bilioneras_in_parallel',exist_ok=True)

tasks = []
for i in range(50):
    output_file = os.path.join('Bilioneras_in_parallel',f'Bilionera {i+1}')
    task = download_song('Bilionera', 'Otilia', output_file)
    tasks.append(task)
result = await asyncio.gather(*tasks) #to stop printing None
```

## Product page html parse

```python
from html_saver import save_page_html,parse_using_ai
```

### Using 8 product pages from different region of the globe

```python
urls = ['https://www.mykitsch.com/products/satin-heatless-curling-set-sunset-tie-dye','https://www.mykitsch.com/products/castor-oil-shampoo-conditioner-bar-combo-2pc','https://www.amazon.com/Beats-Studio-Pro-Personalized-Compatibility/dp/B0C8PT7DF5/','https://www.amazon.com/dp/B09JL41N9C/','https://www.flipkart.com/motorola-amphisoundx-200-w-bluetooth-soundbar/p/itme99fe4fffa928?','https://www.australiathegift.com.au/shop/kangaroo-stick-pins/','https://www.newzealandsouvenirs.com/collections/our-products/products/beer-gear-beer-clothing-speights-beer-medal-t-shirt-trsp60220','https://www.bestbuy.ca/en-ca/category/computers-tablets/20001?icmp=home_shopbycategory_computing']
len(urls)
```

### Running html parsing function asynchronously to save html pages

```python
tasks = []
for idx, url in enumerate(urls):
    output_file = f'new_of{idx+1}.txt'
    task = save_page_html(url,output_file)
    tasks.append(task)
html_txt_files = await asyncio.gather(*tasks)    
```

### Running gemini asynchronously parsing algo for retreiving product info

```python
tasks = []
for idx, file_path in enumerate(html_txt_files):
    task = parse_using_ai(urls[idx],file_path)
    tasks.append(task)
rw = await asyncio.gather(*tasks)    
```
